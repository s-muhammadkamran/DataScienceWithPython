# Apply

It applies a function to each element in a series

```python
#Load this DataFrame from internet and save it to Data Directory
#pd.read_csv('http://bit.ly/kaggletrain').to_csv('Data\\train.csv')
train = pd.read_csv('Data\\train.csv', index_col=0)
train.head()
```

```python
#Apply name method to the all the values of Name columns 
#and store the results in a new column
train["Name_length"] = train.Name.apply(len)
train.loc[0:10, ['Name', 'Name_length']]
```

```python
#Apply numpy ciel function to Fare value

train['Fare_Ciel'] = train.Fare.apply(np.ceil)
train.loc[0:10, ['Fare', 'Fare_Ciel']]
```

```python
#Get the first name from name
train.Name.str.split(',').apply(lambda x: x[0]).head()
```

```python
#Load this DataFrame from internet and save it to Data Directory
#pd.read_csv('http://bit.ly/drinksbycountry').to_csv('Data\\drinksbycountry.csv')
drinks = pd.read_csv('Data\\drinksbycountry.csv', index_col=0)
drinks.head()
```

```python
#Get All rows with columns ranging between beer_servings to wine_servings
#Then apply max function on it, axis 0 means column wise apply i.e. 
#max of each column
drinks.loc[:, 'beer_servings':'wine_servings'].apply(max, axis=0) #Axis=0 means apply in verticaly direction i.e. column wise

#max value of beer_servings column
#max value of spirit_servings column
#max value of wine_servings column
```

```python
#Get All rows with columns ranging between beer_servings to wine_servings
#Then apply max function on it, axis 1 means row wise apply i.e. 
#max of each row
drinks.loc[:, 'beer_servings':'wine_servings'].apply(max, axis=1) #Axis=1 means apply in horizontally direction i.e. row wise

#max value of row 0 is 0
#max value of row 1 is 132
#...
#...
#max value of row n is ...
```

```python
#Get All rows with columns ranging between beer_servings to wine_servings
#Then apply np.argmax function on it, which instead of max value returns the 
#column index of max value. 
#axis 1 means column wise apply i.e. max of each row
drinks.loc[:, 'beer_servings':'wine_servings'].apply(np.argmax, axis=1) #Axis=1 means apply in horizontally direction i.e. row wise

#max of row 0 is at column index 0
#max of row 1 is at column index 1
#...
#...
#max of row n is at column index ...
```

# Map

You can run the map function against the Pandas' Series i.e. specific column of a DataFrame. The map function takes 2 kind of arguments:

1. A mapping function
2. A dictionary

If a mapping function is passed. The map will execute the passed mapping function against each row or in a DataFrame, you can then control what values you can modify in the map funciton.

If a dictionary is passed, then each occurance of the key value in the dictionary will be replaced with the corresponding value in the dictionary.

```python
# Add a new column in DataFrame where sex when female = 0, and male = 1
train['sex_num'] = train.Sex.map({'female':0, 'male':1})
```

```python
# Read first 10 rows and columns Sex and Sex_Num
train.loc[0:10, ['Sex', 'sex_num']]
```

```python
data = pd.DataFrame({"power_level": [25000, 20000, 4000, 1500, 3000, 2000, 1600, 2000],
         "uniform color": ["orange", "blue", "black", "orange", "purple", "green", "orange", "orange"],
         "species": ["saiyan","saiyan","saiyan","half saiyan", "namak","human","human","human"]}, 
         index = ["Goku","Vegeta", "Nappa","Gohan", "Piccolo","Tien","Yamcha", "Krillin"])
data
```

```python
# Use .map() to apply a function to a pandas Series
# Data Frame columns are Series

def my_map_function(x):
    if x > 10000:
        return "High"
    if x > 2000:
        return "Medium"
    else:
        return "Low"
```

```python
# Add the generating series back to DataFrame
data.insert(1, "powers_category", data.power_level.map(my_map_function, na_action='ignore')) #na_action = NAN Action
data
```

```python
# It also maps the series to a key value pair mapping dictionary
# It substitute each occurance of a key with it's value in the series
#To map Series values based on key: value correspondence
# Pass a mapping dictionary to .map()
# Add the generating series back to DataFrame

correspondance = {
    "saiyan":"alien",
    "namak":"alien",
    "human":"earthling",
    "half saiyan":"earthling"
}

data.insert(4, "species_type", data.species.map(correspondance))
data
```

# ApplyMap

This Dataframe method apply the map function to whole dataframe and returns a new dataframe. Map function is applied to Series in a DataFrame, while ApplyMap is applied on the whole DataFrame and returns a DataFrame as well.

```python
df = pd.DataFrame(np.random.randn(4,4))
df
```

```python
def add1(x):
    return x + 1

new_df = df.applymap(add1)
print(type(new_df))
new_df
```

```python
# Same as above, but using anonymous functions.

new_df = df.applymap(lambda x: x+1)
new_df
```
