# **LOC**

Selecting Rows with Labels

Syntax = DataFrame.loc[what_rows_do_I_Want : What_cols_do_I_Want]

ufo.loc[0:2, :] *#means I want rows 0 to 2 with all columns. Unlike index slicing loc is inclusive from both rows and columns*

ufo.loc[0:5, [List of Columns]] *#means first 6 rows with list of columns specified*

```python
ufo.loc[0:2, :]
```

```python
#If you leave out the comma and write [0:2] it means all columns same as above
ufo.loc[0:2]
```

```python
ufo.loc[0:4, ['City', 'State']] #First 5 rows with City and State Columns 
```

```python
ufo.loc[0:, 'City':'State'] #All rows with columns between City and State
```

```python
ufo.loc[ufo.City == 'Oakland', ['State', 'City']] # return all rows where City is Oakland with State and City Columns
```

# ILOC

Selecting rows with Integer Position. I stands for integer

DataFrame.iloc[:, 0:4] # Return all rows with range of columns

```python
ufo.iloc[:, 0:4]
```

```python
ufo.iloc[0:3, :] #First 3 rows with all columns
```

```python
ufo.iloc[0:3, 0:3] #First 3 rows with first 3 columns
```

```python
ufo.iloc[0:3, [0,1,4]]
```
