```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Life expectance with respect to GDP

### Load the data from csv file using Pandas' DataFrame

```python
df = pd.read_csv("Data\\gdp_exp_data.csv")

gdp_cap = df.gdp_cap
life_exp = df.life_exp
population = df.population
colors = df.colors

print(df.head())
print(len(df), len(gdp_cap), len(life_exp), len(population), len(colors))
```

### Generate scatter plot from loaded data

```python
#Set Default Theme of Seaborn for plots
sns.set_theme()

# Define the scatter plot as below. All list values must be scaller and of same length
# x = values for x axis, y = values for y axis, s = size of dots, c = color for each dot, alpha = opacity
plt.scatter(x = gdp_cap, y = life_exp, s = np.array(population) * 2, c = colors, alpha = 0.8)

# Other customizations
plt.xscale('log') # Convert the xscale from normal to logrithmic
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007 - CSV File Data')
plt.xticks([1000,10000,100000], ['1k','10k','100k']) # Labels of ticks at x axis

plt.text(1550, 71, 'India') # Add text to x,y coordinate
plt.text(5700, 80, 'China')

plt.grid(True) # Show gridlines

# Show the plot
plt.show()
```
